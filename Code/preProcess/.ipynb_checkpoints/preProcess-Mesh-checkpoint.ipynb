```python
# scale and rotation scanned meshes

import os
import math
import numpy as np

file_dir = '/home/yan/Data2/NOMO_Project_P2/Dataset/NOMO3D_Dataset2/scans/female_ori/'
out_dir = '/home/yan/Data2/NOMO_Project_P2/Dataset/NOMO3D_Dataset2/scans/female/'

scale_factor = 1000  # unit: m
# scale_factor = 1       # unit: mm

# rotation based x-axis
theta_x = np.radians(-90)
c_x = np.cos(theta_x)
s_x = np.sin(theta_x)

r_x = np.zeros((3, 3), dtype=float)
r_x[0, 0] = 1
r_x[1, 1] = c_x
r_x[1, 2] = -s_x
r_x[2, 1] = s_x
r_x[2, 2] = c_x

# rotation based y-axis
theta_y = np.radians(-90)
c_y = np.cos(theta_y)
s_y = np.sin(theta_y)

r_y = np.zeros((3, 3), dtype=float)
r_y[0, 0] = c_y
r_y[0, 2] = s_y
r_y[1, 1] = 1
r_y[2, 0] = -s_y
r_y[2, 2] = c_y

# rotation based z-axis
theta_z = np.radians(90)
c_z = np.cos(theta_z)
s_z = np.sin(theta_z)

r_z = np.zeros((3, 3), dtype=float)
r_z[0, 0] = c_z
r_z[0, 1] = -s_z
r_z[1, 0] = s_z
r_z[1, 1] = c_z
r_z[2, 2] = 1

obj_list = os.listdir(file_dir)

print len(obj_list)
#------------------------------------------------------------------------
for obj in obj_list:
    vertices = []
    faces = []
    # read and scale vertices and faces
    with open(file_dir + obj, 'r') as fp:
        for line in fp:
            if line[0] == 'v':
                v = map(float, line[2:].split())
                v = np.divide(v, scale_factor)
                vertices.append(v)
            elif line[0] == 'f':
                f = map(int, line[2:].split())  # .obj index starts from 1
                faces.append(f)
                
    # substract mean value           
    v_mean = np.mean(vertices, axis=0)
    vertices = vertices - v_mean
    
    # rotation, based on x and y axis
    v_tran = np.transpose(vertices)
    v_rot = np.dot(r_x, v_tran)
    v_rot = np.dot(r_y, v_rot)
    v_rot = np.transpose(v_rot)
    
    # move the lowest point to the origin
    y_min_index = np.argmin(v_rot, axis=0)
    y_min_value = v_rot[y_min_index[1]][1]
    
    for index in range(0, len(v_rot)):
        v_rot[index][1] = v_rot[index][1] - y_min_value
    
    # save
    with open(out_dir+obj, 'w') as fp:
        for v in v_rot:
            fp.write( 'v %f %f %f\n' % ( v[0], v[1], v[2]) )
        for f in faces: # Faces are 1-based, not 0-based in obj files
            fp.write( 'f %d %d %d\n' %  (f[0], f[1], f[2]) )
            
print 'Done'
```
