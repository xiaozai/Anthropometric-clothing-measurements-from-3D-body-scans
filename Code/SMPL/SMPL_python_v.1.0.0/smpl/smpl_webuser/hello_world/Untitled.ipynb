```python
# step 1: load .obj into SMPL, generate modified .obj
# step 2: with blender, generate .mtl file
# step 3: with pywavefront, display result .obj

import subprocess
import os
import numpy as np

from smpl_webuser.serialization import load_model


import sys
sys.path.append('/home/yan/Data2/3D_Human_Body_Estimation/PyWavefront-master/')
import ctypes

import pyglet
from pyglet.gl import *

import pywavefront


##-----------------  step 1 -----------------------------------##

m = load_model( '../../models/basicModel_f_lbs_10_207_0_v1.0.0.pkl' )

m.pose[:] = np.zeros(72)
m.betas[:] = np.zeros(10)

m.pose[48:51] = [0, 0, -1]
m.pose[51:54] = [0, 0, 1]
## Write to an .obj file
outmesh_path = '/home/yan/Data2/3D_Human_Body_Estimation/AlignmentUI/SMPL_out/test.obj'
with open( outmesh_path, 'w') as fp:
    for v in m.r:
        fp.write( 'v %f %f %f\n' % ( v[0], v[1], v[2]) )

    for f in m.f+1: # Faces are 1-based, not 0-based in obj files
        fp.write( 'f %d %d %d\n' %  (f[0], f[1], f[2]) )
        
##-----------------  step 2 -----------------------------------##
# subprocess.check_output(['blender --background --python /home/yan/Data2/3D_Human_Body_Estimation/obj_to_obj.py'])
os.system('blender --background --python /home/yan/Data2/3D_Human_Body_Estimation/AlignmentUI/obj_to_obj.py')


##-----------------  step 3 -----------------------------------##


rotation = 0

# meshes = pywavefront.Wavefront('zero_pose.obj')
meshes = pywavefront.Wavefront('/home/yan/Data2/3D_Human_Body_Estimation/AlignmentUI/blender_out/test.obj')

window = pyglet.window.Window()

lightfv = ctypes.c_float * 4

@window.event
def on_resize(width, height):
    glMatrixMode(GL_PROJECTION)
    glLoadIdentity()
    gluPerspective(60., float(width)/height, 1., 100.)
    glMatrixMode(GL_MODELVIEW)
    return True

@window.event
def on_draw():
    window.clear()
    glLoadIdentity()

    glLightfv(GL_LIGHT0, GL_POSITION, lightfv(-1.0, 1.0, 1.0, 0.0))
    glEnable(GL_LIGHT0)

    glTranslated(0, 0, -3)
    glRotatef(rotation, 0, 1, 0)
    # glRotatef(-25, 1, 0, 0)
    # glRotatef(45, 0, 0, 1)
    glEnable(GL_LIGHTING)

    meshes.draw()

def update(dt):
    global rotation
    rotation += 90*dt
    if rotation > 720: rotation = 0

pyglet.clock.schedule(update)

pyglet.app.run()
```
